# Word2Vec 만들기

이번 튜토리얼에서는 IMDB 영화리뷰 데이터셋을 가지고 Word Embedding을 해보고자 합니다. Word Embedding Vector를 만들기 위해 다음의 과정을 진행하고자 합니다.

* Data Filtering
* Tokenization
* Gensim 라이브러리를 활용한 Word2Vec 생성

다음의 과정이 끝나게 되면 Word Embedding Vector를 생성하게 됩니다. Word2Vec에서 이 벡터들은 각 단어들 간의 관계를 확인 할 수 있습니다. 모델 생성 후에는 다음의 실습을 해볼 것 입니다. 
* Gensim word2vec 모델에서 지원하는 연산 함수를 활용하여 각 단어들 간의 유사관계를 확인해 볼 것입니다. 
* 또한 시각적으로 단어들 간의 관계를 보기 위해 TSNE로 변환하여 어휘 plot을 보일 것입니다.

## Data Loading

우선 pandas 라이브러리를 활용하여 데이터들을 가져옵니다.

```python
import pandas as pd
from IPython.display import display

train = pd.read_csv("./src/labeledTrainData.tsv", header=0, delimiter="\t", quoting=3)
test = pd.read_csv("./src/testData.tsv", header=0, delimiter='\t', quoting=3)
unlabeled_train = pd.read_csv( "./src/unlabeledTrainData.tsv", header=0, 
 delimiter="\t", quoting=3 )
```

```python
print("Read %d labeled train reviews, %d labeled test reviews, " \
 "and %d unlabeled reviews\n" % (train["review"].size,  
 test["review"].size, unlabeled_train["review"].size ))
```

## Data Filtering 및 Tokenization

텍스트 데이터 내용에서 불필요한 요소들을 제거하고 word token 단위로 데이터를 생성할 수 있는 함수를 구현합니다.

```python
from bs4 import BeautifulSoup
import re
from nltk.corpus import stopwords

def review_to_wordlist( review, remove_stopwords=False ):
    review_text = BeautifulSoup(review, 'html.parser').get_text()
    review_text = re.sub("[^a-zA-Z]"," ", review_text)
    words = review_text.lower().split()
    if remove_stopwords:
        stops = set(stopwords.words("english"))
        words = [w for w in words if not w in stops]
    return(words)
```

```python
display(review_to_wordlist(train['review'][0], True))
```

## Sentence Tokenization

데이터에 따라서 텍스트가 장문일 수 있습니다. 따라서 문장단위로 토크나이즈를 해줘야 할 상황도 생깁니다. 여기선, nltk 라이브러리를 활용하여 문장을 구분하는 함수를 구현합니다.

```python
import nltk.data

tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')

def review_to_sentences(review, tokenizer, remove_stopwords=False):
    raw_sentences = tokenizer.tokenize(review.strip())
    sentences = []
    for raw_sentence in raw_sentences:
        if len(raw_sentence) > 0:
            sentences.append(review_to_wordlist( raw_sentence, remove_stopwords))
    return sentences
```

```python
sentences = []  

print("Parsing sentences from training set")
for review in train["review"]:
    sentences += review_to_sentences(review, tokenizer)

print("Parsing sentences from unlabeled set")
for review in unlabeled_train["review"]:
    sentences += review_to_sentences(review, tokenizer)
```

```python
print(len(sentences))
for i in range(3):
    print(sentences[i])
```

## Word2Vec 모델 생성

gensim 라이브러리를 활용하여 모델을 생성하도록 합니다. 생성하기 위해서는 몇 가지의 파라메터들을 입력해야 합니다.

* num_features: 단어의 feature 갯수
* min_word_count: 최소 count 단어에 대한 제한 (n개 미만의 count가 발생한 단어는 모델에 반영하지 않는다.)
* num_workers: 모델 트레이닝 시 사용하는 쓰레드 갯수
* context: 윈도우 사이즈
* downsampling: 빈도가 있는 단어에 대한 다운샘플링 정도

```python
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s',\
    level=logging.INFO)

# Set values for various parameters
num_features = 300    # Word vector dimensionality                      
min_word_count = 40   # Minimum word count                        
num_workers = 4       # Number of threads to run in parallel
context = 10          # Context window size                                                                                    
downsampling = 1e-3   # Downsample setting for frequent words

# Initialize and train the model (this will take some time)
from gensim.models import word2vec
print("Training model...")
model = word2vec.Word2Vec(sentences, workers=num_workers, \
            size=num_features, min_count = min_word_count, \
            window = context, sample = downsampling)

# If you don't plan to train the model any further, calling 
# init_sims will make the model much more memory-efficient.
model.init_sims(replace=True)
```

## 모델 저장

```python
# It can be helpful to create a meaningful model name and 
# save the model for later use. You can load it later using Word2Vec.load()
model_name = "300features_40minwords_10context"
model.save(model_name)
```

### Word Vector를 Plot에 그려보자

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.manifold import TSNE

wv, vocabulary = model.wv.syn0, model.wv.index2word

fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 80.0
fig_size[1] = 50.0

tsne = TSNE(n_components=2, random_state=0)
np.set_printoptions(suppress=True)
Y = tsne.fit_transform(wv[:1000,:])

plt.scatter(Y[:, 0], Y[:, 1])
for label, x, y in zip(vocabulary, Y[:, 0], Y[:, 1]):
    plt.annotate(label, xy=(x, y), xytext=(0, 0), textcoords='offset points')
plt.show()
```
